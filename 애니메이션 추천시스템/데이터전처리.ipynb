```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('최종데이터.csv', na_values =['0'])
data
```

### 태그 전처리  
수상경력이 있는 경우, 태그가 밀려서 2개의 열로 나오기 때문에 2개의 태그 열을 합치는 작업을 해줘야한다.

```python
tag_1 = data[['태그']]
list_tag_1 = tag_1.values.tolist()
```

```python
s_tag_1 = [0]*2120

for cnt in range(len(list_tag_1)):
    s_tag_1[cnt] = ''.join(str(list_tag_1[cnt]))

for i in range(len(list_tag_1)):
    if s_tag_1[i][0:4] != "['태그":
        s_tag_1[i] = ""
    else :
        s_tag_1[i] = s_tag_1[i].replace('[','')
        s_tag_1[i] = s_tag_1[i].replace("'",'')
        s_tag_1[i] = s_tag_1[i].replace("]",'')
        s_tag_1[i] = s_tag_1[i].replace("태그\\n",'')
```

```python
tag_2 = data[['태그2']]
list_tag_2 = tag_2.values.tolist()
```

```python
s_tag_2 = [0]*2120

for cnt2 in range(len(list_tag_2)):
    s_tag_2[cnt2] = ''.join(str(list_tag_2[cnt2]))

for j in range(len(list_tag_2)):
    if s_tag_2[j][0:4] != "['태그":
        s_tag_2[j] = ""
    else:
        s_tag_2[j] = s_tag_2[j].replace('[','')
        s_tag_2[j] = s_tag_2[j].replace("'",'')
        s_tag_2[j] = s_tag_2[j].replace("]",'')
        s_tag_2[j] = s_tag_2[j].replace("태그\\n",'')
```

```python
data_tag1 = pd.DataFrame(list(zip(s_tag_1)), columns = ["태그"])
data_tag2 = pd.DataFrame(list(zip(s_tag_2)), columns = ["태그2"])
result = pd.concat([data_tag1,data_tag2],axis=1)
```

```python
result['태그완성'] = result['태그'] + " " + result['태그2']
result_mod = result.drop([result.columns[0],result.columns[1]], axis=1)
```

```python
final = pd.concat([data,result_mod],axis=1)
final
```

### 제작정보 열 분리하기  
제작정보 열에서 제작, 그림, 출시 등의 정보를 나누어 3열로 분리

```python
final['제작'] = final['제작정보'].str.split('\n').str[0]
final['각본'] = final['제작정보'].str.split('\n').str[1]
final['그림'] = final['제작정보'].str.split('\n').str[2]
final['출시'] = final['제작정보'].str.split('\n').str[3]

final
```

```python
#제작 컬럼에서 '그림'을 가지고 있는 행 인덱스
final['제작'] = final['제작'].fillna('')

gurim = np.where(final['제작'].str.contains('그림'))

#제작 컬럼에서 '각본'을 가지고 있는 행 인덱스 

gakbon = np.where(final['제작'].str.contains('각본'))

#제작 컬럼에서 '출시'을 가지고 있는 행 인덱스 

chulsi = np.where(final['제작'].str.contains('출시'))
```

```python
#제작 컬럼에 '그림'이 들어간 행 정보 저장 및 그림 컬럼에 옮겨서 저장.
jeajak_gurim = []*2120

jeajak_gurim = final['제작'][gurim[0]]
```

```python
#제작 컬럼에 '각본'이 들어간 행 정보 저장 및 그림 컬럼에 옮겨서 저장.

jeajak_gakbon = []*2120

jeajak_gakbon = final['제작'][gakbon[0]]
```

```python
#제작 컬럼에 '출시'가 들어간 행 정보 저장 및 그림 컬럼에 옮겨서 저장.
jeajak_chulsi = []*2120
jeajak_chulsi = final['제작'][chulsi[0]]
```

```python
#각본 컬럼에서 '그림'을 가지고 있는 행 인덱스
final['각본'] = final['각본'].fillna('')

gurim2 = np.where(final['각본'].str.contains('그림'))

#각본 컬럼에서 '출시'을 가지고 있는 행 인덱스

chulsi2 = np.where(final['각본'].str.contains('출시'))
```

```python
#각본 컬럼에 '그림'이 들어간 행 정보 저장
gakbon_gurim = []*2120
gakbon_gurim = final['각본'][gurim2[0]]


#각본 컬럼에 '출시'가 들어간 행 정보 저장
gakbon_chulsi = []*2120
gakbon_chulsi = final['각본'][chulsi2[0]]
```

```python
#그림 컬럼에서 '출시'을 가지고 있는 행 인덱스
final['그림'] = final['그림'].fillna('')

chulsi3 = np.where(final['그림'].str.contains('출시'))
```

```python
#그림 컬럼에 '출시'가 들어간 행 정보 저장
gurim_chulsi = []*2120
gurim_chulsi = final['그림'][chulsi3[0]]
```

```python
#제작 컬럼 내용 옯겨서 저장
final['그림'][gurim[0]] = jeajak_gurim
final['각본'][gakbon[0]] = jeajak_gakbon
final['출시'][chulsi[0]] = jeajak_chulsi 

#각본 컬럼 내용 옮겨서 저장
final['그림'][gurim2[0]] = gakbon_gurim
final['출시'][chulsi2[0]] = gakbon_chulsi

#그림 컬럼 내용 옮겨서 저장
final['출시'][chulsi3[0]] = gurim_chulsi
```

```python
#제작 컬럼에 '그림', '출시', '각본'이 들어간 행 내용 삭제
final['제작'][gurim[0]] = None
final['제작'][gakbon[0]] = None
final['제작'][chulsi[0]] = None
```

```python
#각본에서 '그림', '출시'가 들어간 행 삭제
gak_gurim = np.where(final['각본'].str.contains('그림'))
final['각본'][gak_gurim[0]] = None
gak_chulsi = np.where(final['각본'].str.contains('출시'))
final['각본'][gak_chulsi[0]] = None
guri_chulsi = np.where(final['그림'].str.contains('출시'))
final['그림'][guri_chulsi[0]] = None
```

```python
final
```

```python
def cleaning(x) :
  if str(x).find("제작 : ") != -1 :
    return str(x).replace("제작 : ","") 
  return str(x)
def cleaning2(x) :
  if str(x).find("그림 : ") != -1 :
    return str(x).replace("그림 : ","") 
  return str(x)
def cleaning3(x) :
  if str(x).find("출시 : ") != -1 :
    return str(x).replace("출시 : ","") 
  return str(x)
def cleaning4(x) :
  if str(x).find("각본 : ") != -1 :
    return str(x).replace("각본 : ","") 
  return str(x)
```

```python
final['제작']=final['제작'].apply(cleaning)
final['각본']=final['각본'].apply(cleaning4)
final['그림']=final['그림'].apply(cleaning2)
final['출시']=final['출시'].apply(cleaning3)
final
```

```python
rd_loc = 0 

final_df = pd.DataFrame(columns = ['제목', '장르', '줄거리','평점', '태그'])

for i , row in final.iterrows():
    title = row['제목'] 
    genre = row['장르']
    story = row['줄거리']
    rate = row['평점']
    
    for s in row['태그완성'].split('#'):
        s = str(s).strip() 
        final_df.loc[rd_loc] = [title, genre, story, rate, s] 
        rd_loc = rd_loc + 1
```
